##### **Ecosystem:**
In natural sciences, ‘ecosystems’ are generally defined as a system, or a group of interconnected elements, formed by the interaction of a community of organisms with their environment. 

##### **Startup:**
A startup or start-up is a company or project undertaken by an entrepreneur to seek, develop, and validate a scalable business model. Startups are new businesses that intend to grow large beyond the solo founder. At the beginning, startups face high uncertainty and have high rates of failure, but a minority of them do go on to become successful and influential.

##### **India:** 
India is a country that occupies the greater part of South Asia. India is made up of 28 states and eight union territories, and its national capital is New Delhi. It is the seventh-largest country by area and the most populous country as of June 2023.

We need to combine category with similar labels

```python
#funcition to combine similar labels, this funciton will recieve the labels and the matching percent, default percent is 80
def combinedSimilarLabels(labelSeries, threshold=80):

    label={}

    for i in labelSeries.unique():
        matches = process.extract(i, labelSeries.unique(), limit=None)

        close_Matches = [j[0] for j in matches if j[1] >= threshold]

        combinedLabel = ', '.join(close_Matches)
        
        label[i] = combinedLabel

    combinedSeries = labelSeries.map(label)

    return combinedSeries

        
```

```python
df_all['Sector'] = combinedSimilarLabels(df_all['Sector'], threshold=95)
```

##### **QUESTION 1: What is the Total Value of the Indian Startup Ecosystem?** 

**1-a. How Many Startups were Founded between 2018 to 2021?**

```python
#checking number of companies founded before and after 2018

before2018=0
after2018=0

unique_companies = df_all.groupby('Company').first().reset_index()

for x in unique_companies['Founded']:
    if x>= 2018 and x<=2021:
        after2018 += 1
    elif x != 0:
        before2018 += 1

print(f"Total number of companies: {(before2018+after2018)}")
print(f"Total number of companies before 2018: {before2018}")
print(f"Total number of companies between 2018-2021: {after2018}")
print(f"Percent of companies in this period: {(after2018/(before2018+after2018))*100:.2f}%")
```

```python
palette = sns.color_palette("viridis", 10)
```

```python
#checking the trend of startups founded over time
(unique_companies
 .groupby('Founded')
 .Founded
 .count()
 .plot(kind='bar',
       figsize= (10,5),
       title = 'Trend of Startups Founded over the years',
       color=palette
       )
)
```

* The trend of startups shows that there has been increasing number of startups founded over the years and the ecosystem is growing with 46.34% of the startups founded post 2018.
